### 4.1 Finish the __generator__ such that it splits a string into words and conditionally produces words from that string based on their length (5)
- The generator takes 2 arguments: *sentence* and *n*
 - sentence is the string to split
 - n is the maximum number of characters the words it produces should have
  - the __default__ value for n should be 3
  
Iterating over this:  
wordProducer("This is a sentence with a few words but it does not mean anything")  
should produce this:  

---
is  
a  
a  
few  
but  
it  
not    

---
and Iterating over this:  
wordProducer("This is a sentence with a few words but it does not mean anything",2)  
should produce this: 

---
is  
a  
a  
it  

```python
def wordProducer(sentence, n=3):
    
    # Split the sentence into words and loop through each word
    # Append words to desired list where length <= n
    desired = [i for i in sentence.split() if len(i) <= n]

    # Yield results
    for i in desired:
        yield(i)


# Call generator
ThreeCharWords = wordProducer("This is a sentence with a few words but it does not mean anything")
# Loop through results
print("Results for ThreeCharWords:")
for res in ThreeCharWords:
    print(res)

    
# Call generator    
TwoCharWords = wordProducer("This is a sentence with a few words but it does not mean anything", 2)
# Loop through results
print("\n\nResults for TwoCharWords:")
for res in TwoCharWords:
    print(res)
```

# 8. Natural Language Processing

### Start by executing the cell below to load the data
*You will have to have an active internet connection as it uses requests to fetch a book from the Gutenberg library*

```python
# These are the imports that you might need for this section
# if you want to use others add those imports in the cell you
# intend to use them and a short comment why.
import nltk
import requests
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# This is the book ALICE’S ADVENTURES IN WONDERLAND from the Gutenberg library
# the preface has been removed for you but the Chapter titles all remain
alice = requests.get("https://www.gutenberg.org/files/11/11-0.txt").content.decode('utf8')[1492:].replace("\r","")
print(alice)
```

### 8.1 Split the text into paragraphs, remove all newline characters, and remove Chapter headings (10)
- paragraphs are separated by double newline characters \n\n
- Chapter headings start with an all caps CHAPTER

__Print the first paragraph only__ which should be:

*Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do: once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, “and what is the use of a book,” thought Alice “without pictures or conversations?”*

```python
# Replace chapter seperator with \n\n and then split into paragraphs as per \n\n paragraph seperator
paragraphs = alice.replace('\n\n\n', '\n\n').split('\n\n')
# Remove \n and replace with space
paragraphs = [i.replace('\n',' ') for i in paragraphs]
# Remove chapter headers
paragraphs = [i for i in paragraphs if i[0:7] != 'CHAPTER']

# Display first paragraph
display(paragraphs[0])
```

### 8.2 Counting characters and words (5)
- What is the average character length of the paragraphs
- How many words to the paragraphs contain on average (assume words are separated by any whitespace character)  

__round the mean to 2 decimal places__

The output should be:

--

The mean character length of paragraphs is  XXXX  
The mean number of words of the paragraphs is  YYYY

--

with __XXXX__ and __YYYY__ replaced with the actual calculated mean values

```python
# Import regex to assist in cleaning
import re

# Convert all sentence enders to full-stop ( ? ! .. ) for consistency in future paragraph splitting
paragraphs = [re.sub('\?|\!|\'\..\'', '.', i) for i in paragraphs]
# Convert all intra-sentence grammar to comma ( : ; ) for consistency in future paragraph splitting
paragraphs = [re.sub('\:|\;', ',', i) for i in paragraphs]
# Remove special characters that could be considered superfluous: _ ( ) * ” “ ‘ — -
paragraphs = [re.sub('_|\(|\)|\*|\”|\“|\‘|\—|\-', '', i) for i in paragraphs]
# Replace multiple white spaces with single white space
paragraphs = [re.sub('\s+', ' ', i) for i in paragraphs]
# Trim beginning and end of sentences
paragraphs = [i.strip() for i in paragraphs]
# Remove any empty entries
paragraphs = [i for i in paragraphs if i != '']

# Calculate the length of each paragraph
para_len = [len(i) for i in paragraphs]

# Calculate the mean character length of paragraphs
print(f"The mean character length of paragraphs is: {round(sum(para_len)/len(para_len), 2)}")

# Split paragraphs into lists of words within sentences and remove blank entries
para_words = [re.split('\s+', i) for i in paragraphs]
para_words = [i for i in para_words if i != '']

word_count = [len(i) for i in para_words]
print(f"The mean number of words per paragraph is: {round(sum(word_count)/len(word_count), 2)}")

# Check first paragraph and previously troublesome paragraph for content sanity check
print("\nThe first paragraph:\n", para_words[0])
print("\nA cleaned paragraph:\n", para_words[26])   # Contained white space, brackets and quotes
```

### 8.3 Plot the distribution of the compound sentiment score of each sentence in the Alice text (10)
Create a graph that is as close as possible to the accompanying figure (sentiment_dist.png)
- You will have to tokenize the text into sentences (*use nltk tokenization tools*)
- and score each sentence separately using the VADER SentimentIntensityAnalyzer tool
- Do not save the figure, __just display it in the notebook inline__
- The figure was created using the seaborn library which was imported above

Here are some of the properties you might need
- Axis label fontsize is 16
- Title fontsize is 20
- Tick label fontsize is 16
- Figure size is a square 12 inches



```python
analyser = SentimentIntensityAnalyzer()

# Convert previously cleaned paragraphs into sentences
sentences = nltk.sent_tokenize(' '.join(paragraphs))

# Added after new information received on Teams
# Check NLTK version
print(f"NLTK version: {nltk.__version__}")

# Calcuate scores per sentence
scores = [analyser.polarity_scores(i)['compound'] for i in sentences]
# Version 3.5 causes huge spike in zero value scores, excluding these improves graph shape
scores = [i for i in scores if i != 0.0]

# Attempt to plot seaborn chart
sns.set(rc={'figure.figsize':(12, 12)})
sns.set_style("white")
plot = sns.distplot(scores, bins=11, kde=True, rug=True, label="Sentiment score")
plot.set_title('The distribution of the compound sentiment score', size=20)
plot.legend(loc='upper left', fontsize=16)
plot.set_xlabel('Compound Score', size=16)
plot.set(xlim = (-1.5, 1.5))
plot.set_ylabel('Frequency', size=16)
```
